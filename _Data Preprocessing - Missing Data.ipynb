#### Impute missing values for categorical variable ####

```python
df
```

```python
df.isnull().sum()
```

```python
df.groupby('Street')['LotFrontage'].mean()
```

```python
df.groupby('LandContour')['LotFrontage'].mean()
```

```python
df['LotFrontage'].fillna(df.groupby('Street')['LotFrontage'].transform('mean'))
```

```python
df['Street'].fillna(df['Street'].mode())
```
